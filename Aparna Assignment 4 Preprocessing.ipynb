### Filling the missing values present in the dataset

```python
# Three columns contain missing values
# Age (float),Cabin and Embarked(object)
```

```python
# Age (numerical)
```

```python
plt.hist(data['Age'])
plt.show()
```

```python
# to fill missing values,Median methond can be used
```

```python
age_med=data['Age'].median()
data['Age']= data['Age'].fillna(age_med)
```

```python
data.isna().sum()
```

```python
# to fill missing values in cabin and Embarked column, mode method can be used
```

```python
cabin_mode=data['Cabin'].mode()[0]
data['Cabin']=data['Cabin'].fillna(cabin_mode)
```

```python
data.isna().sum()
```

```python
# embarked 
```

```python
embar_mode=data['Embarked'].mode()[0]
data['Embarked']=data['Embarked'].fillna(embar_mode)
```

```python
data.isna().sum()
```
